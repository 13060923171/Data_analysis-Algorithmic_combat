# 1、实验背景

时间序列分析是一种处理时间序列数据或趋势分析的统计技术。时间序列数据是指数据处于一系列特定的时间段或间隔内。数据可以分为三种类型：
* 时间序列数据：对变量在不同时间所取值的一组观察。
* 横截面数据：在同一时间点收集的一个或多个变量的数据。
* 合并数据：时间序列数据和横截面数据的组合。

时间序列数据由四个部分组成，为了让这个概念更加清晰，这里对时间序列的各个组成部分进行了直观的解释。：
* 趋势成分：这是一种在很长一段时间内以合理可预测的模式向上或向下移动的变化。
* 季节性成分：是有规律和周期性的变化，并在特定时间段内重复，例如一天、一周、一个月、一个季节等
* 周期性成分：是与商业或经济“繁荣-萧条”周期相对应或遵循其自身特殊周期的变化
* 随机分量：是不稳定或残差的变异，不属于上述三类中的任何一类。

### 4.2.2 数据预处理

#### 4.2.2.2 特征工程
几乎每个时间序列问题都会有一些外部特征或一些内部特征工程来帮助建立模型。

让我们添加一些基本特征，例如广泛用于时间序列问题的可用数字特征的滞后值。由于我们需要预测一天的股票价格，我们不能使用同一天的特征值，因为它们在实际推理时间是不可用的。我们需要使用统计数据，例如其滞后值的均值、标准差。

我们将使用三组滞后值，一组前一天，一组表示前一周，另一组表示前一个月。

```python
#特征生成
df.reset_index(inplace=True)
lag_features = ["High", "Low", "Volume", "Turnover", "Trades"]
window1 = 3
window2 = 7
window3 = 30

df_rolled_3d = df[lag_features].rolling(window=window1, min_periods=0)
df_rolled_7d = df[lag_features].rolling(window=window2, min_periods=0)
df_rolled_30d = df[lag_features].rolling(window=window3, min_periods=0)

df_mean_3d = df_rolled_3d.mean().shift(1).reset_index().astype(np.float32)
df_mean_7d = df_rolled_7d.mean().shift(1).reset_index().astype(np.float32)
df_mean_30d = df_rolled_30d.mean().shift(1).reset_index().astype(np.float32)

df_std_3d = df_rolled_3d.std().shift(1).reset_index().astype(np.float32)
df_std_7d = df_rolled_7d.std().shift(1).reset_index().astype(np.float32)
df_std_30d = df_rolled_30d.std().shift(1).reset_index().astype(np.float32)

for feature in lag_features:
    df[f"{feature}_mean_lag{window1}"] = df_mean_3d[feature]
    df[f"{feature}_mean_lag{window2}"] = df_mean_7d[feature]
    df[f"{feature}_mean_lag{window3}"] = df_mean_30d[feature]
    
    df[f"{feature}_std_lag{window1}"] = df_std_3d[feature]
    df[f"{feature}_std_lag{window2}"] = df_std_7d[feature]
    df[f"{feature}_std_lag{window3}"] = df_std_30d[feature]

df.fillna(df.mean(), inplace=True)

df.set_index("Date", drop=False, inplace=True)
df.head()
```

* 对于boosting模型，添加日期时间特征（如小时、日、月）非常有用，适用于提供有关数据中时间组件的模型信息。
* 对于时间序列模型，虽然没有明确要求提供此信息，但我们可以这样做，以便在完全相同的一组特征上比较所有模型。

```python
df.Date = pd.to_datetime(df.Date, format="%Y-%m-%d")
df["month"] = df.Date.dt.month
df["week"] = df.Date.dt.week
df["day"] = df.Date.dt.day
df["day_of_week"] = df.Date.dt.dayofweek
df.drop(['Deliverable Volume','%Deliverble'],axis=1,inplace=True)
df.head()
```

接下来将数据拆分成训练集、验证集以及特征值
* **train**： 2008年5月26日至2018年12月31日的数据。
* **valid**： 2019年1月1日至2019年12月31日的数据。
* 提供给时间序列问题的附加特征称为外生回归量。

```python
df_train = df[df.Date < "2019"]
df_valid = df[df.Date >= "2019"]

exogenous_features = ["High_mean_lag3", "High_std_lag3", "Low_mean_lag3", "Low_std_lag3",
                      "Volume_mean_lag3", "Volume_std_lag3", "Turnover_mean_lag3",
                      "Turnover_std_lag3", "Trades_mean_lag3", "Trades_std_lag3",
                      "High_mean_lag7", "High_std_lag7", "Low_mean_lag7", "Low_std_lag7",
                      "Volume_mean_lag7", "Volume_std_lag7", "Turnover_mean_lag7",
                      "Turnover_std_lag7", "Trades_mean_lag7", "Trades_std_lag7",
                      "High_mean_lag30", "High_std_lag30", "Low_mean_lag30", "Low_std_lag30",
                      "Volume_mean_lag30", "Volume_std_lag30", "Turnover_mean_lag30",
                      "Turnover_std_lag30", "Trades_mean_lag30", "Trades_std_lag30",
                      "month", "week", "day", "day_of_week"]
```

### 4.3.1 Auto ARIMAX
ARIMA模型根据给定的时间序列自身的过去值，即其自身的滞后和滞后的预测误差来解释给定的时间序列，因此该模型可用于预测未来值。

ARIMA模型需要某些输入参数：p代表 AR(p)，q代表MA(q)，d代表I(d)。值得庆幸的是，有一个可以自动选择这些参数的模型，被称为Auto ARIMA。当外生回归量与 ARIMA 一起使用时，它通常被称为 ARIMAX。

```python
#构建模型
model = auto_arima(df_train.VWAP, exogenous=df_train[exogenous_features], trace=True, error_action="ignore", suppress_warnings=True)
#模型训练
model.fit(df_train.VWAP, exogenous=df_train[exogenous_features])
```

```python
#模型预测
forecast = model.predict(n_periods=len(df_valid),exogenous=df_valid[exogenous_features])
df_valid["Forecast_ARIMAX"] = forecast
```

```python
#模型评估
print("RMSE of Auto ARIMAX:", np.sqrt(mean_squared_error(df_valid.VWAP, df_valid.Forecast_ARIMAX)))
print("\nMAE of Auto ARIMAX:", mean_absolute_error(df_valid.VWAP, df_valid.Forecast_ARIMAX))
```

```python
#模型可视化
df_valid[["VWAP", "Forecast_ARIMAX"]].plot(figsize=(14, 7))
```

【结果分析】

Auto ARIMAX 模型在根据前一天的数据预测股票价格方面似乎做得相当好。其他模型能打败它吗？

### 4.3.2 Prophet模型
Prophet 是 Facebook 开发的开源时间序列模型。它于 2017 年初发布。 摘自主页：
>Prophet 是一种基于可加模型预测时间序列数据的程序，其中非线性趋势与每年、每周和每天的季节性以及假日效应相吻合。它最适用于具有强烈季节性影响和多个季节历史数据的时间序列。Prophet 对缺失数据和趋势变化具有稳健性，并且通常可以很好地处理异常值。

Prophet遵循 sklearn 模型的API。我们创建了一个Prophet类的实例，然后调用它的fit和predict方法。

Prophet的输入始终是具有两列的数据框：ds 和 y。ds (datestamp) 列应该是Pandas期望的格式，最好是 YYYY-MM-DD 表示日期或 YYYY-MM-DD HH:MM:SS 表示时间戳。y 列必须是数字，表示我们希望预测的度量。

```python
#模型实例化
model_fbp = Prophet()
for feature in exogenous_features:
    model_fbp.add_regressor(feature)
#训练模型
model_fbp.fit(df_train[["Date", "VWAP"] + exogenous_features].rename(columns={"Date": "ds", "VWAP": "y"}))
#模型预测
forecast = model_fbp.predict(df_valid[["Date", "VWAP"] + exogenous_features].rename(columns={"Date": "ds"}))
df_valid["Forecast_Prophet"] = forecast.yhat.values
```

```python
#模型评估
print("RMSE of Auto ARIMAX:", np.sqrt(mean_squared_error(df_valid.VWAP, df_valid.Forecast_ARIMAX)))
print("RMSE of Prophet:", np.sqrt(mean_squared_error(df_valid.VWAP, df_valid.Forecast_Prophet)))
print("\nMAE of Auto ARIMAX:", mean_absolute_error(df_valid.VWAP, df_valid.Forecast_ARIMAX))
print("MAE of Prophet:", mean_absolute_error(df_valid.VWAP, df_valid.Forecast_Prophet))
```

```python
#模型可视化
df_valid[["VWAP", "Forecast_ARIMAX", "Forecast_Prophet"]].plot(figsize=(14, 7))
```

【结果分析】

在这种情况下，Auto ARIMAX 的性能优于 Prophet。

### 4.3.3 LightGBM模型
时间序列问题通常被转换为表格结构，并输入到LightGBM和XGBoost等增强模型中。时间序列中数据点的顺序方面存在信息丢失，但可以通过日期时间数据以在一定程度上规避这一问题。

请注意，本实验使用默认参数，后续可以调整它们以改善结果。

```python
#数据准备
params = {"objective": "regression"}
dtrain = lgb.Dataset(df_train[exogenous_features], label=df_train.VWAP.values)
dvalid = lgb.Dataset(df_valid[exogenous_features])
#模型训练
model_lgb = lgb.train(params, train_set=dtrain)
#模型预测
forecast = model_lgb.predict(df_valid[exogenous_features])
df_valid["Forecast_LightGBM"] = forecast
```

```python
#模型评估
print("RMSE of Auto ARIMAX:", np.sqrt(mean_squared_error(df_valid.VWAP, df_valid.Forecast_ARIMAX)))
print("RMSE of Prophet:", np.sqrt(mean_squared_error(df_valid.VWAP, df_valid.Forecast_Prophet)))
print("RMSE of LightGBM:", np.sqrt(mean_squared_error(df_valid.VWAP, df_valid.Forecast_LightGBM)))
print("\nMAE of Auto ARIMAX:", mean_absolute_error(df_valid.VWAP, df_valid.Forecast_ARIMAX))
print("MAE of Prophet:", mean_absolute_error(df_valid.VWAP, df_valid.Forecast_Prophet))
print("MAE of LightGBM:", mean_absolute_error(df_valid.VWAP, df_valid.Forecast_LightGBM))
```

```python
#模型可视化对比
df_valid[["VWAP", "Forecast_ARIMAX", "Forecast_Prophet", "Forecast_LightGBM"]].plot(figsize=(14, 7))
```

【结果分析】

LightGBM 的表现非常差！这是对时间序列使用boosting模型的一个关键的问题。请记住，boosting模型仅限于在训练数据中出现的目标值范围内进行预测。

那么为什么boosting方法仍然如此流行呢？因为boosting模型仅在数据趋势非常强的情况下表现不佳。股票价格就是一个例子，它通常具有很强的趋势。

# 5、总结分析
* Auto ARIMAX 是一个很好的模型，但是像 Facebook 的 Prophet 这样的新算法也非常强大，并且变得越来越智能。不要害怕尝试新技术。
* 设置合适的验证框架极其重要。能够尝试和试验各种模型并客观地比较它们。
* 基于滞后的特征在提供有关时间序列数据的趋势信息方面非常有用。
* 外生回归器有助于提供有关时间序列的外部信息。它们在大多数模型中往往非常重要。
* 像 LightGBM 这样的 Boosting 模型被限制在训练数据中目标变量的值范围内进行预测，并且不会在有强烈趋势时进行推断。
