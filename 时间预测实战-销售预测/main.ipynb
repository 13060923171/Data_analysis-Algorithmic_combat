```python
#导入处理数据的库
import pandas as pd
import numpy as np
```

```python
#整理总数据
df_total = pd.read_excel("Table 89 3.xlsx",sheet_name='总体').loc[:,['Unnamed: 0','百万港元','指数','按年变动百分率','指数.1','按年变动百分率.1']]
#把N.A.替换成nan
df_total = df_total.replace('N.A.',np.nan)
#把nan所在的那一行全部删掉
df_total.dropna(axis=0,how='any',inplace=True)
df_total
```

```python
#清洗数据
def dispose_df(year):
    df_total = pd.read_excel("Table 89 3.xlsx", sheet_name='{}'.format(year)).loc[:,
               ['Unnamed: 0', '百万港元', '指数', '按年变动百分率', '指数.1', '按年变动百分率.1']]
    #和上面一样将数据替换
    df_total = df_total.replace('N.A.', np.nan)
    #删除
    df_total.dropna(axis=0, how='any', inplace=True)
    def double_df(x):
        return str(year) + "-" + str(x)
    #生成新的一列数据
    df_total['Unnamed: 0'] = df_total['Unnamed: 0'].apply(double_df)
    
    return df_total

```

```python
frames = []
#获取每个列表中数据
for i in range(2004,2022):
    df = dispose_df(i)
    frames.append(df)
#将多个dataframe合并
result = pd.concat(frames)
#重新生成columns
result.columns = ['year','money','index','variation','index2','variation2']
#将时间作为表格的索引
result.year = pd.to_datetime(result.year)
result.index = result.year
result
```

```python
#生成偏自相关图(PACF)
fig = plt.figure(figsize=(12,8))
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(data,lags=40,ax=ax2)
```

```python
from itertools import product
import warnings
from tqdm import tqdm
warnings.filterwarnings('ignore')
#设置参数范围
ps = range(0,10)
qs = range(0,10)
parameters = product(ps,qs)
parameters_list = list(parameters)


#寻找最优arma模型参数，即best_aic最小
results = []
#正无穷
best_aic = float('inf')

#去测试，寻找最优模型
for param in tqdm(parameters_list):
    try:
        model = ARMA(data,order=(param[0],param[1])).fit()
    except:
#         print('参数错误:',param)
        continue
    #一个比对的过程，选出最优的模型
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param,model.aic])

#输出最优模型
result_table = pd.DataFrame(results)
result_table.columns = ['parameters','aic']
print('最优模型:',best_model.summary())
```

```python
#在选出模型检验之后，对arma模型所产生的残差做相关图
resid = best_model.resid
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid,lags=40,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid,lags=40,ax=ax2)
```

```python
#基本符合正态分布
#生成正态图
from statsmodels.graphics.api import qqplot
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
fig = qqplot(resid, line='q', ax=ax, fit=True)
```

```python
#用Ljung-box检验:检验结果就是看最后一列前十二行的检验概率
#prob值均小于0.05，所以残差序列存在自相关性
r,q,p = sm.tsa.acf(resid.squeeze(),qstat=True)
data1 = np.c_[range(1,41),r[1:],q,p]
table = pd.DataFrame(data1,columns=['lag','AC','Q','Prob(>Q)'])
table.set_index('lag')
```
