```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_excel('data.xls',sheet_name='内蒙古')
data2 = pd.read_excel('data.xls',sheet_name='四川')
data3 = pd.read_excel('data.xls',sheet_name='甘肃')
data4 = pd.read_excel('data.xls',sheet_name='青海')
data5 = pd.read_excel('data.xls',sheet_name='宁夏')
```

```python
data = pd.concat([data1,data2,data3,data4,data5])
data = data.drop(['Unnamed: 0'],axis=1)
```

```python
data.isnull().sum()
```

```python
data.dropna(how='any',inplace=True)
```

```python
data.head()
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.subplots(figsize = (20,9))
sns.heatmap(data.corr(), annot=True,cmap="PiYG")
```

```python
flue_gas_train_x = data[['电耗（亿千瓦时）','电力热力的生产和供应业（2）','工业环境治理投资额（万元）','能耗（万吨）','工业固体废物利用量（万吨）','工业固体废弃物（万吨）']]
flue_gas_train_y = data['废气（亿立方米）']
```

```python
from sklearn import preprocessing
from sklearn.metrics import r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
```

```python
# data1.dropna(how='any',inplace=True)
# flue_gas_test_x = data1[['电耗（亿千瓦时）','电力热力的生产和供应业（2）','工业环境治理投资额（万元）','能耗（万吨）','工业固体废物利用量（万吨）','工业固体废弃物（万吨）']]
# flue_gas_test_y = data1['废气（亿立方米）']
train_x,test_x,train_y,test_y = train_test_split(flue_gas_train_x,flue_gas_train_y,test_size=0.2)
```

```python
ss = preprocessing.StandardScaler()
train_ss_x = ss.fit_transform(train_x)
test_ss_x = ss.transform(test_x)
```

```python
#创建一个KNN
knn = KNeighborsRegressor()
knn.fit(train_ss_x,train_y)
predict_y = knn.predict(test_ss_x)
print('KNN准确率:%.2lf' % r2_score(test_y,predict_y))
```

```python
corr = data.corr()
```

```python
def gx(zd):
    a = corr.loc[(corr['{}'.format(zd)] > 0.9) & (corr['{}'.format(zd)] != 1) ,'{}'.format(zd)]
    b = list(a.index)
    flue_gas_train_x = data[b]
    flue_gas_train_y = data['{}'.format(zd)]
    train_x,test_x,train_y,test_y = train_test_split(flue_gas_train_x,flue_gas_train_y,test_size=0.2)
    return train_x,test_x,train_y,test_y,b
    
```

```python
train_x,test_x,train_y,test_y,b = gx('工业固体废物利用量（万吨）')
ss = preprocessing.StandardScaler()
train_ss_x = ss.fit_transform(train_x)
test_ss_x = ss.transform(test_x)
#创建一个KNN
knn = KNeighborsRegressor()
knn.fit(train_ss_x,train_y)
predict_y = knn.predict(test_ss_x)
print('KNN准确率:%.2lf' % r2_score(test_y,predict_y))
```
